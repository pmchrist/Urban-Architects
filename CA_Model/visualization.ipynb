```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('./results/result.csv')
# print(df.head(5))
print(df.columns)
# print(df.iloc[:, 1])
for i in range(1, 3):
    plt.plot(df.iloc[:, i], label=df.columns[i])
plt.legend()
plt.savefig('./results/climate.png')
plt.show()
for i in range(3, len(df.columns)):
    plt.plot(df.iloc[:, i], label=df.columns[i])
plt.legend()
plt.savefig('./results/mean.png')
plt.show()
```

```python
for i in range(1, len(df.columns)):
    plt.hist(df[df.columns[i]], label = df.columns[i])
    # plt.xlabel = df.columns[i]
    plt.legend()
    plt.savefig('./results/distribution'+df.columns[i]+'.png')
    plt.show()
```

```python
import seaborn as sns
ax = sns.boxplot(data=df.iloc[:,1:3])
# ax = sns.boxplot(x='categories', y='oxygen', hue='target', data=df)
ax.set_xticklabels(ax.get_xticklabels(),rotation=15)
plt.show()

ax = sns.boxplot(data=df.iloc[:,3:])
# ax = sns.boxplot(x='categories', y='oxygen', hue='target', data=df)
ax.set_xticklabels(ax.get_xticklabels(),rotation=15)
plt.show()
```
